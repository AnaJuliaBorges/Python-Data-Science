```
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (30, 16))
```

```
dados = pd.read_csv('aluguel_residencial_tratado_sem_outliers.csv', sep = ';')
dados.head()
```

```
area = plt.figure()
g1 = area.add_subplot(2, 2, 1) #Isso quer dizer que dentro dessa figura há um gráfico, duas linhas e duas colunas na posição 1
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```
g1.scatter(dados.Valor, dados.Area) #um gráfico de dispersão
g1.set_title('Valor X Área')

g2.hist(dados.Valor) #gráfico de distribuição de frequências
g2.set_title('Histograma')

dados_g3 = dados.Valor.sample(100) #uma amostra aleatória dentro do próprio dataframe
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor Médio por Tipo')
```
