##Imóveis residenciais

```
import pandas as pd
```

```
properties = pd.read_csv('aluguel.csv', sep = ';')
properties
```

```
list(properties['Tipo'].drop_duplicates())
```

```
types_residential = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Flat',
 'Casa de Vila',
 'Loft',
 'Studio']
```

```
selection_residential = properties['Tipo'].isin(types_residential)
selection_residential
```

```
properties_residential = properties[selection_residential]
properties_residential
```

```
list(properties_residential['Tipo'].drop_duplicates())
```

```
properties_residential.index = range(properties_residential.shape[0])
properties_residential
```

##Exportando arquivo

```
properties_residential.to_csv('aluguel_residencial.csv', sep = ';')
```
