##Criando novas variáveis

```
import pandas as pd
```

```
residentials = pd.read_csv('aluguel_residencial_tratado.csv', sep = ';')
```

```
residentials['Valor Bruto'] = residentials['Valor'] + residentials['Condominio'] + residentials['IPTU']
residentials
```

```
residentials['Valor m2'] = residentials['Valor'] / residentials ['Area']
residentials
```

```
residentials['Valor m2'] = residentials['Valor m2'].round(2)
```

```
residentials['Valor Bruto m2'] = (residentials['Valor Bruto'] / residentials['Area']).round(2)
residentials                                  
```

```
# http://www.dsc.ufcg.edu.br/~pet/jornal/maio2013/materias/tutoriais.html - Lambda
```

```
residentials['Tipo'].drop_duplicates()
```

```
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```
#apply() - Aplica uma função no eixo determinado do dataframe, padrão eixo x(0)
```

```
residentials['Tipo Agregado'] = residentials['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
residentials
```

##Excluindo variáveis

```
dados_ux = pd.DataFrame(residentials[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
dados_ux
```

```
del dados_ux['Valor Bruto m2']
dados_ux
```

```
dados_ux.pop('Tipo Agregado')
dados_ux
```

```
residentials.drop(['Valor Bruto m2', 'Valor Bruto'], axis = 1, inplace=True)
residentials
```
