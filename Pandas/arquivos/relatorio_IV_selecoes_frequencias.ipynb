```
import pandas as pd
```

```
residentials = pd.read_csv('aluguel_residencial.csv', sep = ';')
residentials
```

```
#Selecione somente os imóveis classificados com tipo 'Apartamento'
```

```
selection_apartment = residentials['Tipo'] == "Apartamento"
residentials[selection_apartment]
```

```
apartment = residentials[selection_apartment].shape[0]
apartment
```

```
#Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'
```

```
selection_types = (residentials['Tipo'] == 'Casa') | (residentials['Tipo'] == 'Casa de Condomínio') | (residentials['Tipo'] == 'Casa de Vila')
residentials[selection_types]
```

```
types = residentials[selection_types].shape[0]
types
```

```
#Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites
```

```
selection_areas = (residentials['Area'] >= 60) & (residentials['Area'] <= 100)
residentials[selection_areas]
```

```
areas =residentials[selection_areas].shape[0]
areas
```

```
#Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00
```

```
selection_room_rent = (residentials['Quartos'] >= 4) & (residentials['Valor'] < 2000)
residentials[selection_room_rent].head(10)
```

```
room_rent = residentials[selection_room_rent].shape[0]
room_rent
```

```
print(f"Nº de imóveis classificados com tipo 'Apartamento' -> {apartment}")
print(f"Nº de imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'-> {types}")
print(f"Nº de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites -> {areas}")
print(f"Nº de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00 -> {room_rent}")
```
