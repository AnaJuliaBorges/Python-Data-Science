##Tratamento de dados faltantes

```
import pandas as pd
```

```
residentials = pd.read_csv('aluguel_residencial.csv', sep = ';')
residentials
```

```
residentials.isnull()
```

```
residentials.notnull()
```

```
residentials.info()
```

```
residentials[residentials['Valor'].isnull()]
```

```
residentials.dropna(subset = ['Valor'], inplace=True)
```

```
residentials[residentials['Valor'].isnull()]
```

##Tratando condominio, valor e IPTU

```
residentials[residentials['Condominio'].isnull()].shape[0]
```

```
selection = (residentials['Tipo'] == 'Apartamento') & (residentials['Condominio'].isnull())
residentials[selection].shape[0]
```

```
residentials = residentials[~selection]
```

```
residentials[selection].shape[0]
```

```
residentials[residentials['Condominio'].isnull()].shape[0]
```

```
selection = (residentials['Condominio'].isnull()) | (residentials['IPTU'].isnull())
residentials[selection]
```

```
residentials = residentials.fillna({'Condominio': 0, 'IPTU': 0})
residentials
```

```
residentials.info()
```

```
residentials.to_csv('aluguel_residencial_tratado.csv', sep= ';', index=False)
```
